```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import time
```

```python
headers = {'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.03987.162 Safari/537.36'}
Player_name, Age, Price, Clubs, Position, Nationality = [], [], [], [], [], []

for page in range(1,26):
    print(f"Scraping page {page}...")
    url = f"https://www.transfermarkt.com/premier-league/marktwertaenderungen/wettbewerb/GB1/pos//detailpos/0/verein_id/0/land_id/0/plus/1/sort/marktwert.desc/page/{page}" 
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.text, 'html.parser')

    for tag in soup.find_all('td', class_='hauptlink'):
            a_tag = tag.find('a')
            if a_tag and '/profil/spieler/' in a_tag.get('href'):
                Player_name.append(a_tag.text.strip())

    for tag in soup.find_all('td',class_='zentriert'):
        text = tag.get_text(strip=True)
        if text.isdigit():
            age = int(text)
            if 15 <= age <= 45:
                Age.append(age)

    for tag in soup.find_all('td', class_='rechts hauptlink'):
        Price.append(tag.text.strip())

    for tag in soup.find_all('td', class_='zentriert'):
        a_tag = tag.find('a')
        if a_tag and '/startseite/verein/' in a_tag.get('href'):
            title = a_tag.get('title')
            if title:
                Clubs.append(title.strip())

    for row in soup.find_all('tr'):
        tds = row.find_all('td')
        if len(tds) == 1 and not tds[0].find('a') and not tds[0].find('img'):
            Position.append(tds[0].text.strip())

    for tag in soup.find_all('td', class_='zentriert'):
        img_tag = tag.find('img')
        if img_tag and 'flaggenrahmen' in img_tag.get('class'):
            title = img_tag.get('title')
            if title:
                Nationality.append(title.strip())
            

    time.sleep(1)



# Align all lists
min_len = min(len(Player_name), len(Age), len(Price), len(Clubs), len(Position), len(Nationality))
Player_name = Player_name[:min_len]
Age = Age[:min_len]
Price = Price[:min_len]
Clubs = Clubs[:min_len]
Position = Position[:min_len]
Nationality = Nationality[:min_len]

# Save to CSV
df = pd.DataFrame({
    "Player Name": Player_name,
    "Age": Age,
    "Market Value": Price,
    "Club": Clubs,
    "Position": Position,
    "Nationality": Nationality
})

df.to_csv("premier_league_players.csv", index=False)
print("✅ CSV saved successfully!")
```
